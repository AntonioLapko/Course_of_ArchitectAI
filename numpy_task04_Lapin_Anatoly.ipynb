###Задача № 1

```python
import numpy as np
A = np.matrix([[1, 2, 1],
               [-2, 1, 1]])
A = A.transpose()

B = np.matrix([[3, 1],
               [-1, 0],
               [2, 4]])

C = np.matrix([[-2, 2],
               [1, -1],
               [1, 2]])

print('A+2B = \n', A+2*B)
print('B+2C = \n', B+2*C)
```

### Задача № 3

```python
import numpy as np
a = np.linspace(0, 1.4, 15)
b = np.linspace(1, 6, 6)
a = a.reshape(5, 3)
b = b.reshape(3, 2)
print('Первая матрица 5х3\n', a)
print('Вторая матрица 3х2\n', b)
print('Перемножение первой и второй матриц\n', a.dot(b))
```

###Задача № 4

```python
import numpy as np
a = np.arange(start=1, stop=72, step=2, dtype='int32')
b = a%3
c = []
for i in range(0, len(b)):
  if b[i] == 0:
    c.append(a[i])
print('Исходный вектор\n',a)
print('Вектор с кратными трем значениями исходного вектора\n',c)
```
