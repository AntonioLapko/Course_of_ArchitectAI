##Решение через numpy

```python
import numpy as np
print("Task 1")
A = np.matrix([6, -8, 3])
B = np.matrix([[1],
               [1],
              [4]])
print("A=\n",A)
print("B=\n",B)
C = A*B
print("A*B=\n",C)
print("\nTask 2")
D = np.matrix([[-3, -7],
              [-10, 6],
              [0, -6],
              [-1, -3]])
E = np.matrix([[-8, 9],
              [4, -2]])
print("D=\n",D)
print("E=\n",E)
F = D*E
print("D*E=\n",F)
```

##Решение через циклы

```python
rwsA = int(input ("Введите количество строк матрицы А: "))
print("rows A =",rwsA)
clsA = int(input ("Введите количество столбцов матрицы А: "))
print("colums B =", clsA)

print("Заполните матрицу А")
A = []
for i in range(rwsA):
    print("Введите через пробел значения столбцов строки", i+1)
    A.append(list(map(int, input().split())))

print("Матрица A:")
for i in range(rwsA):
    for j in range(clsA):
        print(A[i][j], end=' ')
    print()

rwsB = int(input ("Введите количество строк матрицы В: "))
print("rows B =",rwsB)
clsB = int(input ("Введите количество столбцов матрицы В: "))
print("colums B =",clsB)

print("Заполните матрицу B")
B = []
for i in range(rwsB):
    print("Введите через пробел значения столбцов строки", i+1)
    B.append(list(map(int, input().split())))

print("Матрица B:")
for i in range(rwsB):
    for j in range(clsB):
        print(B[i][j], end=' ')
    print()

C = [[0 for row in range(clsB)] for col in
range(rwsA)]

for i in range(rwsA):
  for j in range(clsB):
    for z in range(clsA):
      C[i][j] += A[i][z] * B[z][j]

print("Результат перемножения матриц A и B: ")
for i in range(len(C)):
    for j in range(len(C[0])):
        print(C[i][j], end=' ')
    print()

```
